```python
# url -> text and images from message div
# images from div -> hyperlink -> request -> downloaded image in os.listdir()

def url_to_transcript(url):
    outter = []
    page = requests.get(url).content
    soup = BeautifulSoup(page, "html.parser")
    for message in soup.findAll('div', {'class': ['message']}):
        inner = [] 
        for p_img in message.find_all(['p', 'img']):
            if p_img.name == 'img':
                inner.append(p_img['data-src'])
            if p_img.name == 'p':
                inner.append(p_img.text)
        outter.append(inner)
    print(url)
    return outter

    
def message_to_text(url):
    page = requests.get(url).content
    soup = BeautifulSoup(page, 'html.parser')
    for message in soup.find_all('div', {'class': ['message']}):
        p_text = [p.text for p in message.find_all('p')]
        images = [img['data-src'] for img in message.findAll('img')]

    print(url)
    return p_text + images
```
